## Linear Regression

Este estimador es una regresión lineal simple conocida como **Ordinary Least Square - OLS**. 

En *Scikit-learn* se requiere que la variable objetivo ***y*** tenga una sola dimensión, mientras que los ejemplos de entrenamiento ***(X)*** deben presentarse como un array de dos dimensiones. En los datos que se generaron antes el valor de X tiene las siguientes dimensiones:

```python
x.shape
```

Por lo tanto, para no tener problemas con la API de *Scikit-learn* se debe convertir en un array de dos dimensiones.

```python
X = X[:, np.newaxis]
```

Al revisar nuevamente las dimensiones, se obtiene:

```python
X.shape
```

**Creación del Modelo**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.25, train_size=0.75, random_state=123)

LinearR =  LinearRegression()
LinearR.fit(X_train, y_train)
```

Después de ajustar el modelo a los valores presentado en la fase de entrenamiento, se obtienen los siguientes parámetros:

```python
print("Pendiente: ", LinearR.coef_)
print("Punto de intercepción: ", LinearR.intercept_)
```

$$y = Pendiente * X + PuntoIntercepción$$

La representación gráfica de los datos y el modelo de ajuste es:

```python
min_pt = X.min() * LinearR.coef_[0] + LinearR.intercept_
max_pt = X.max() * LinearR.coef_[0] + LinearR.intercept_

plt.plot([X.min(), X.max()], [min_pt, max_pt])
plt.plot(X_train, y_train, 'o', color='red');
```

Ahora se usará le método *predict* para obtener las predicciones del modelo con los datos de *test*.

```python
y_pred_train = LinearR.predict(X_train)
y_pred_test = LinearR.predict(X_test)
```

Para los datos X_train se espera que los datos esten dentro de la línea generada por los parámetros obtenidos del modelo. El modelo logra obtener el comportamiento de la pendiente general de los datos, pero no los detalles de ajuste. 

```python
plt.plot(X_train, y_train, 'o', label="datos", color="red")
plt.plot(X_train, y_pred_train, 'o', label="predicciones", color="lightblue")
plt.plot([X.min(), X.max()], [min_pt, max_pt], label='ajuste', color="blue")
plt.legend(loc='best');
```

Las predicciones para el conjunto *test* (nunca visto antes por el modelo) deben ajustarse igualmente a la línea generada por el modelo.

```python
plt.plot(X_test, y_test, 'o', label="datos", color="red")
plt.plot(X_test, y_pred_test, 'o', label="predicción", color="lightblue")
plt.plot([X.min(), X.max()], [min_pt, max_pt], label='ajuste', color='blue')
plt.legend(loc='best');
```

Para evaluar cuantitativamente los estimadores de regresión, *Scikit-learn* usa al igual que con los estimadores de clasificación, el método *score*. Mientras que para las clasificadores *score* proporciona el porcentaje de aciertos correctamente clasificados ***CCR***, para los regresores usa el ***Coeficiente de Determinacion - $R^2$***.

Para el caso de Regresión Lineal Simple de este ejemplo, $R^2$ es el cuadrado del ***Coeficiente de correlación de Pearson***. Otra forma de evaluar al modelo es mediante el uso de ***Mean Squared Error (MSE)*** (Diferencia al cuadrado entre los valores predichos y los valores reales).

$$MSE = \frac{1}{n}\sum_{i=1}^n(ValorPredicho_{i}-ValorReal_{i})^{2} $$

Entre más cerca a 0 se encuentre el valor del MSE, indica un mejor ajuste del modelo.

```python
LinearR.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test, y_pred_test)
print(MSE)
```

## KNeighbors Regression

Aunque este método es más popular en clasificación que en regresión, se muestra como ejemplo para contrastar con el modelo de Regresión Lineal.

```python
from sklearn.neighbors import KNeighborsRegressor

NeighborsR = KNeighborsRegressor()
NeighborsR.fit(X_train, y_train)
```

```python
y_pred_train = NeighborsR.predict(X_train)
y_pred_test = NeighborsR.predict(X_test)
```

```python
plt.plot(X_train, y_train, 'o', label="datos", markersize=10, color='blue')
plt.plot(X_train, y_pred_train, 's', label="predicción", markersize=4, color='red')
plt.legend(loc='best');
```

```python
plt.plot(X_test, y_test, 'o', label="datos", markersize=8, color='blue')
plt.plot(X_test, y_pred_test, 's', label="predicción", markersize=4, color='red')
plt.legend(loc='best');
```

El estimador Kneighbors se ajusta mejor a los datos que el modelo lineal. La puntuación $R^2$ es:

```python
NeighborsR.score(X_test, y_test)
```

El valor del MSE es:

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test, y_pred_test)
print(MSE)
```

Al comparar los resultados obtenidos, se observa que el estimador de *Kneighbors* se ajusta mucho mejor a los datos presentads al modelo que el modelo *Linear Regression*.
